```python
import nltk
import re
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

#### Cleaning

```python
paragraph_copy=paragraph
paragraph_copy=re.sub(r'[^\w\s]','',paragraph_copy)
paragraph_copy=re.sub(r'[^\D]','',paragraph_copy)
```

```python
print(paragraph_copy)
```

### Word Tokenization

```python
words=nltk.word_tokenize(paragraph_copy)
```

```python
words
```

### Removing Stop words

```python
stop_words=stopwords.words('english')
print(stop_words)
```

```python
word_filtered=[]
for word in words:
    word=word.lower()
    if word not in stop_words:
        word_filtered.append(word)
```

```python
word_filtered
```

```python
print(f"Number of words before removing stopwords: {len(words)} \nNumber of words after removing stopwords: {len(word_filtered)}")
```

### Lemmatization

```python
lemma_words=[]
lemmatizer=WordNetLemmatizer()
for word in word_filtered:
    lemma_words.append(lemmatizer.lemmatize(word))
```

```python
print(lemma_words)
```

# Bag of Words Implementation

```python
sentences=nltk.sent_tokenize(paragraph)
sentences = [re.sub('[^a-zA-Z ]', '', sentence) for sentence in sentences]
for i in range(len(sentences)):
    words=sentences[i].split()
    for j in range(len(words)):
        if words[j].lower() not in stop_words:
            words[j]=words[j].lower()
        else:
            words[j]=''
    sentences[i]=' '.join(words)
sentences
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
vectorized = vectorizer.fit_transform(sentences)
```
